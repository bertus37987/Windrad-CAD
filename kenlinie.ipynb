```python
import pandas as pd
import plotly.express as px

g = 9.81

df = pd.DataFrame({
    "Typ": ["Rotor", "Rotor", "Rotor", "Rotor", "Pumpe", "Pumpe", "Pumpe"],
    "Masse_g": [10, 200, 500, 700, 200, 500, 700],
    "Zeit_s": [None, None, None, None, 12.0, 6.0, 3.5],
    "P_mW": [8, 69, 143, 131, 27, 69, 60]
})

df["M_Nm"] = (df["Masse_g"] / 1000) * 9.81
df = df.sort_values(["Typ", "M_Nm"])

fig = px.line(df, x="M_Nm", y="P_mW", color="Typ", markers=True)
fig.update_xaxes(title_text="M Rotor (Nm)")
fig.update_yaxes(title_text="Leistung (mW)")
fig.show()
```
