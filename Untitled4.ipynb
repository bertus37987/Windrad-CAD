```python
from os import linesep
import pandas as pd
import plotly.express as px

# Konstanten
g = 9.81
h = 0.80

# Rotor-Daten
rotor = pd.DataFrame({
    "Typ": ["Rotor", "Rotor", "Rotor", "Rotor"],
    "Lastmasse_g": [10, 200, 500, 700],
    "Zeit_s": [18.8, 22.6, 27.5, 42.0]
})
rotor["Lastmasse_kg"] = rotor["Lastmasse_g"] / 1000
rotor["M_N"] = rotor["Lastmasse_kg"] * g
rotor["Leistung_mW"] = rotor["Lastmasse_kg"] * g * h / rotor["Zeit_s"] * 1000

# Pumpen-Daten
pumpe = pd.DataFrame({
    "Typ": ["Pumpe", "Pumpe", "Pumpe"],
    "Lastmasse_g": [200, 500, 700],
    "Zeit_s": [12.0, 6.0, 3.5],
    "Wasser_ml": [27, 69, 60]
})
pumpe["Wassermasse_kg"] = pumpe["Wasser_ml"] / 1000
pumpe["M_N"] = (pumpe["Lastmasse_g"] / 1000) * g
pumpe["Leistung_mW"] = pumpe["Wassermasse_kg"] * g * h / pumpe["Zeit_s"] * 1000

# Beide zusammen
kennlinie = pd.concat([
    rotor[["Typ", "Lastmasse_g", "Zeit_s", "M_N", "Leistung_mW"]],
    pumpe[["Typ", "Lastmasse_g", "Zeit_s", "M_N", "Leistung_mW"]]
], ignore_index=True)

print(kennlinie)

# Plot
fig = px.line(
    kennlinie,
    x="M_N",
    y="Leistung_mW",
    color="Typ",
    markers=True,
    title="Kennlinie von Rotor und Pumpe"
)
fig.update_xaxes(title="Lastkraft M (N)")
fig.update_yaxes(title="Leistung P (mW)")
fig.update_traces(marker=dict(size=14))
fig.show()
```
